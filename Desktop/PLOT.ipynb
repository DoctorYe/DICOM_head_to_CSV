```python
import pandas as pd
from sklearn.manifold import TSNE
Raw_Feature_1045=pd.read_csv('./raw_feature_1045.csv')
tsne = TSNE(n_components=2, random_state=0)
Raw_Feature_1045_2D = tsne.fit_transform(Raw_Feature_1045)
%matplotlib inline
from matplotlib import pyplot as plt
plt.figure(figsize=(30, 20))
for i in range(len(Raw_Feature_1045_2D)):
    plt.scatter(Raw_Feature_1045_2D[i,0],Raw_Feature_1045_2D[i,1])
plt.show()
```

```python
recurrence_colored=[]
for i in Raw_Feature_1045['recurrence']:
    if i==0:
        recurrence_colored.append('r')
    else:
        recurrence_colored.append('g')
plt.figure(figsize=(30, 20))
for i in range(len(Raw_Feature_1045_2D)):
    plt.scatter(Raw_Feature_1045_2D[i,0],Raw_Feature_1045_2D[i,1],c=recurrence_colored[i])
plt.show()
```

# Dimensional reduce using PCA

```python
from sklearn.decomposition import PCA
pca=PCA(n_components=0.90)
radiomic_feature_reduced=pca.fit_transform(Raw_Feature_1045.drop(labels=['recurrence'],axis=1))
import matplotlib.pyplot as plt
import numpy as np
pca_plot=PCA()
pca_plot.fit(Raw_Feature_1045.drop(labels=['recurrence'],axis=1))
plt.plot(np.cumsum(pca_plot.explained_variance_ratio_))
plt.show()
```
